```python
import os
import pandas as pd
from pydub import AudioSegment
from pydub.utils import which

# Set working directory
os.chdir("/content/drive/MyDrive/AI_ML_SEM_5")

# Configure ffmpeg
AudioSegment.converter = which("ffmpeg")
AudioSegment.ffprobe = which("ffprobe")

print("✅ ffmpeg path:", AudioSegment.converter)

# Load the balanced dataset
EmotionDataset_Balanced = pd.read_csv("CSVs/EmotionDataset_Balanced.csv")

# Directory to save silence-trimmed audio files
location = "EmotionDataset_Silenced/"
os.makedirs(location, exist_ok=True)

def detect_leading_silence(audio, silence_threshold=-50.0, chunk_size=10):
    """Detect leading silence duration (in ms)."""
    trim_ms = 0
    audio_length = len(audio)

    while trim_ms < audio_length:
        chunk_dBFS = audio[trim_ms:trim_ms + chunk_size].dBFS
        if chunk_dBFS > silence_threshold:
            break
        trim_ms += chunk_size

    return trim_ms

name = 0
errors = 0
file_data = []

# Process each audio file
for path, emotion in zip(EmotionDataset_Balanced.Path, EmotionDataset_Balanced.Emotions):
    file_name = f"{location}{name}.wav"

    try:
        audio = AudioSegment.from_file(path, format='wav')
    except:
        audio = AudioSegment.from_file(path)

    start_trim = detect_leading_silence(audio)
    end_trim = detect_leading_silence(audio.reverse())
    duration = len(audio)

    new_audio = audio[start_trim:duration - end_trim]

    if len(new_audio) > 0:
        new_audio.export(file_name, format='wav')
    else:
        audio.export(file_name, format='wav')
        errors += 1
        print(f"{path} is completely silent")

    file_data.append((emotion, file_name))

    print(f"Processed audios: {name}, Errors found: {errors}")
    name += 1

# Save the new dataset with trimmed audio paths
EmotionDataset_Silenced = pd.DataFrame(file_data, columns=['Emotions', 'Path'])
EmotionDataset_Silenced.to_csv("CSVs/EmotionDataset_Silenced.csv", index=False)

print("✅ Silence trimming completed!")
print(f"Total processed: {name}, Errors: {errors}")
```

```python

# Set working directory
os.chdir("/content/drive/MyDrive/AI_ML_SEM_5")

# Load the silenced dataset
EmotionDataset_Silenced = pd.read_csv("CSVs/EmotionDataset_Silenced.csv")

# Directory to save augmented audio files
location = "EmotionDataset_Augmented/"
os.makedirs(location, exist_ok=True)

def add_noise(data):
    """Add random Gaussian noise to audio data."""
    noise_amp = 0.04 * np.random.uniform() * np.amax(data)
    return data + noise_amp * np.random.normal(size=data.shape[0])

def stretch_audio(data, rate=0.8):
    """Stretch or compress the audio (change speed)."""
    return librosa.effects.time_stretch(data, rate=rate)

def change_pitch(data, sampling_rate, n_steps=0.7):
    """Shift the pitch of the audio."""
    return librosa.effects.pitch_shift(data, sr=sampling_rate, n_steps=n_steps)

name = 0
file_data = []

# Process each audio file
for path, emotion in zip(EmotionDataset_Silenced.Path, EmotionDataset_Silenced.Emotions):
    # --- Original ---
    file_name = f"{location}{name}.wav"
    print(f"Processed audios: {name}")

    data, sample_rate = librosa.load(path, sr=None)
    sf.write(file_name, data, sample_rate)
    file_data.append((emotion, file_name))
    name += 1

    # --- Noise ---
    file_name = f"{location}{name}.wav"
    print(f"Processed audios: {name}")

    data_noise = add_noise(data)
    sf.write(file_name, data_noise, sample_rate)
    file_data.append((emotion, file_name))
    name += 1

    # --- Stretch ---
    file_name = f"{location}{name}.wav"
    print(f"Processed audios: {name}")

    data_stretch = stretch_audio(data)
    sf.write(file_name, data_stretch, sample_rate)
    file_data.append((emotion, file_name))
    name += 1

    # --- Pitch Variation ---
    file_name = f"{location}{name}.wav"
    print(f"Processed audios: {name}")

    data_pitch = change_pitch(data, sample_rate)
    sf.write(file_name, data_pitch, sample_rate)
    file_data.append((emotion, file_name))
    name += 1

# Save the final augmented dataset
EmotionDataset_Augmented = pd.DataFrame(file_data, columns=["Emotions", "Path"])
EmotionDataset_Augmented.to_csv("CSVs/EmotionDataset_Augmented.csv", index=False)

print("✅ Audio augmentation completed successfully!")
print(f"Total processed files: {name}")
```

```python
import pandas as pd

# Load dataset
EmotionDataset = pd.read_csv("CSVs/EmotionDataset_Augmented.csv")

# Shuffle
EmotionDataset = EmotionDataset.sample(frac=1).reset_index(drop=True)

# Number of emotions
emotions = EmotionDataset["Emotions"].unique()
num_emotions = len(emotions)   # should be 8

# Percentages
p_train = 0.80
p_val   = 0.10
p_test  = 0.10

# DataFrames for splits
Audios_train = pd.DataFrame(columns=["Emotions", "Path"])
Audios_val   = pd.DataFrame(columns=["Emotions", "Path"])
Audios_test  = pd.DataFrame(columns=["Emotions", "Path"])

# Split per emotion
for emotion in emotions:

    subset = EmotionDataset[EmotionDataset["Emotions"] == emotion]

    total = len(subset)
    train_n = int(total * p_train)
    val_n   = int(total * p_val)
    test_n  = total - train_n - val_n   # Remaining → ensures no overlap

    # Assign splits
    train_split = subset.iloc[:train_n]
    val_split   = subset.iloc[train_n : train_n + val_n]
    test_split  = subset.iloc[train_n + val_n :]

    Audios_train = pd.concat([Audios_train, train_split])
    Audios_val   = pd.concat([Audios_val, val_split])
    Audios_test  = pd.concat([Audios_test, test_split])

# Shuffle each split
Audios_train = Audios_train.sample(frac=1).reset_index(drop=True)
Audios_val   = Audios_val.sample(frac=1).reset_index(drop=True)
Audios_test  = Audios_test.sample(frac=1).reset_index(drop=True)

# Save CSVs
Audios_train.to_csv("CSVs/EmotionDataset_div_train.csv", index=False)
Audios_val.to_csv("CSVs/EmotionDataset_div_val.csv", index=False)
Audios_test.to_csv("CSVs/EmotionDataset_div_test.csv", index=False)

# ------- Print stats -------
print("\n✅ Training set created!")
print(Audios_train["Emotions"].value_counts())
print("Shape:", Audios_train.shape)

print("\n✅ Validation set created!")
print(Audios_val["Emotions"].value_counts())
print("Shape:", Audios_val.shape)

print("\n✅ Test set created!")
print(Audios_test["Emotions"].value_counts())
print("Shape:", Audios_test.shape)
```

```python

import cv2
import matplotlib.pyplot as plt
from scipy.io import wavfile

# Directory to save spectrogram images
location = 'Features/Images/'

# Create directory if it doesn't exist
os.makedirs(location, exist_ok=True)

# --- Generate and Save Spectrogram ---
def graph_spectrogram(audio_path, file_name):
    rate, data = wavfile.read(audio_path)
    fig, ax = plt.subplots(1)
    fig.set_size_inches(0.64, 0.64)
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1)
    ax.axis('off')
    try:
        # For mono audio
        ax.specgram(x=data, Fs=rate, cmap='gray_r', noverlap=384, NFFT=512)
    except:
        # For stereo audio (averaging both channels)
        ax.specgram(x=(data[:, 0] + data[:, 1]) / 2, Fs=rate, cmap='gray_r', noverlap=384, NFFT=512)
    fig.savefig(file_name, dpi='figure')
    plt.close(fig=fig)


# --- Prepare Image Dataset from Audio Files ---
def prepare_images(emotions, paths, dataset_name):
    data = []
    labels = []
    img_index = 0

    for idx, audio_path in enumerate(paths):
        file_name = f'{location}{dataset_name}_{img_index}.png'
        graph_spectrogram(audio_path, file_name)

        image = cv2.imread(file_name, cv2.IMREAD_GRAYSCALE)

        # Check if image was loaded correctly
        if image is None:
            print(f"Error loading image for audio: {audio_path}")
            continue

        print(f"Processing image: {idx}")

        data.append(image)
        labels.append(emotions[idx])
        img_index += 1

    # Convert lists to arrays and normalize images
    data = np.array(data, dtype="float") / 255.0
    labels = np.array(labels)

    # Save images and labels as .npy files
    np.save(f'Features/{dataset_name}_images_data.npy', data)
    np.save(f'Features/{dataset_name}_images_labels.npy', labels)


# --- Load Train, Validation, and Test CSVs ---
data_path_Train = pd.read_csv('CSVs/EmotionDataset_div_train.csv')
data_path_Val = pd.read_csv('CSVs/EmotionDataset_div_val.csv')
data_path_Test = pd.read_csv('CSVs/EmotionDataset_div_test.csv')

# --- Generate and Save Spectrogram Images ---
prepare_images(data_path_Train['Emotions'], data_path_Train['Path'], 'train')
prepare_images(data_path_Val['Emotions'], data_path_Val['Path'], 'val')
prepare_images(data_path_Test['Emotions'], data_path_Test['Path'], 'test')
```

```python


# --- Feature Extraction Function ---
def extract_features(audio_data, sample_rate):
    features = np.array([])

    # Zero Crossing Rate (ZCR)
    zcr = np.mean(librosa.feature.zero_crossing_rate(y=audio_data).T, axis=0)
    features = np.hstack((features, zcr))

    # Chroma STFT
    stft = np.abs(librosa.stft(audio_data))
    chroma_stft = np.mean(librosa.feature.chroma_stft(S=stft, sr=sample_rate).T, axis=0)
    features = np.hstack((features, chroma_stft))

    # MFCC (Mel Frequency Cepstral Coefficients)
    mfcc = np.mean(librosa.feature.mfcc(y=audio_data, sr=sample_rate).T, axis=0)
    features = np.hstack((features, mfcc))

    # Root Mean Square Energy (RMS)
    rms = np.mean(librosa.feature.rms(y=audio_data).T, axis=0)
    features = np.hstack((features, rms))

    # Mel Spectrogram
    mel = np.mean(librosa.feature.melspectrogram(y=audio_data, sr=sample_rate).T, axis=0)
    features = np.hstack((features, mel))

    return features


# --- Load and Extract Features from an Audio File ---
def get_features(file_path):
    data, sample_rate = librosa.load(file_path)
    features = extract_features(data, sample_rate)
    return np.array(features)


# --- Process and Save Audio Features ---
def prepare_audios(emotions, paths, dataset_name):
    data = []
    labels = []

    for idx, audio_path in enumerate(paths):
        try:
            feature = get_features(audio_path)
            data.append(feature)
            labels.append(emotions[idx])
            print(f"Processing audio: {idx}")
        except:
            print(f"Error processing audio file: {audio_path}")

    # Convert lists to arrays
    data = np.array(data)
    labels = np.array(labels)

    # Save features and labels as .npy files
    np.save(f'Features/{dataset_name}_audio_data.npy', data)
    np.save(f'Features/{dataset_name}_audio_labels.npy', labels)


prepare_audios(data_path_Train['Emotions'], data_path_Train['Path'], 'train')
prepare_audios(data_path_Val['Emotions'], data_path_Val['Path'], 'val')
prepare_audios(data_path_Test['Emotions'], data_path_Test['Path'], 'test')

# --- Load Dataset Paths ---
data_path_Train = pd.read_csv('CSVs/EmotionDataset_div_train.csv')
data_path_Val = pd.read_csv('CSVs/EmotionDataset_div_val.csv')
data_path_Test = pd.read_csv('CSVs/EmotionDataset_div_test.csv')
```

```python

import cv2
import matplotlib.pyplot as plt
from scipy.io import wavfile

# Directory to save spectrogram images
location = 'Features/Images/'

# Create directory if it doesn't exist
os.makedirs(location, exist_ok=True)

# --- Generate and Save Spectrogram ---
def graph_spectrogram(audio_path, file_name):
    rate, data = wavfile.read(audio_path)
    fig, ax = plt.subplots(1)
    fig.set_size_inches(0.64, 0.64)
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1)
    ax.axis('off')
    try:
        # For mono audio
        ax.specgram(x=data, Fs=rate, cmap='gray_r', noverlap=384, NFFT=512)
    except:
        # For stereo audio (averaging both channels)
        ax.specgram(x=(data[:, 0] + data[:, 1]) / 2, Fs=rate, cmap='gray_r', noverlap=384, NFFT=512)
    fig.savefig(file_name, dpi='figure')
    plt.close(fig=fig)


# --- Prepare Image Dataset from Audio Files ---
def prepare_images(emotions, paths, dataset_name):
    data = []
    labels = []
    img_index = 0

    for idx, audio_path in enumerate(paths):
        file_name = f'{location}{dataset_name}_{img_index}.png'
        graph_spectrogram(audio_path, file_name)

        image = cv2.imread(file_name, cv2.IMREAD_GRAYSCALE)

        # Check if image was loaded correctly
        if image is None:
            print(f"Error loading image for audio: {audio_path}")
            continue

        print(f"Processing image: {idx}")

        data.append(image)
        labels.append(emotions[idx])
        img_index += 1

    # Convert lists to arrays and normalize images
    data = np.array(data, dtype="float") / 255.0
    labels = np.array(labels)

    # Save images and labels as .npy files
    np.save(f'Features/{dataset_name}_images_data.npy', data)
    np.save(f'Features/{dataset_name}_images_labels.npy', labels)


# --- Load Train, Validation, and Test CSVs ---
data_path_Train = pd.read_csv('CSVs/EmotionDataset_div_train.csv')
data_path_Val = pd.read_csv('CSVs/EmotionDataset_div_val.csv')
data_path_Test = pd.read_csv('CSVs/EmotionDataset_div_test.csv')

# --- Generate and Save Spectrogram Images ---
prepare_images(data_path_Train['Emotions'], data_path_Train['Path'], 'train')
prepare_images(data_path_Val['Emotions'], data_path_Val['Path'], 'val')
prepare_images(data_path_Test['Emotions'], data_path_Test['Path'], 'test')
```
